## MCDM Implementation

```python
dict1 = {1:[1,1,1] , 2:[1,2,3] , 3:[2,3,4] , 4:[3,4,5] , 5:[4,5,6] , 6: [5,6,7] , 7:[6,7,8],8:[7,8,9],9:[9,9,9]}
```

```python
terms = 3
```

```python
def AHP_MCDM(criteria):
  data = criteria
  n = len(criteria)
  print("Length is :",n)

  new_matrix = numpy.zeros((n,n,3))
  #print(new_matrix)

  for x in range(n):
    for y in range(n):
      if(data[x][y] >= 1):
        new_matrix[x][y] = dict1[data[x][y]]
      else:
        temp = dict1[(1/data[x][y])]
        for i in range(3):
          new_matrix[x][y][i] = 1.0/temp[2-i]
  return new_matrix
  
  
```

```python
# Calculate Fuzzy Geometric MEAN:
def fuzzy_GM(new_matrix):
  n = len(new_matrix)
  #print(n)

  # take a multi D array and initilize it with 1
  geometric_mean = numpy.ones((n, 3))
  #print(geometric_mean)
  #(l1 * l2, m1 * m2, u1 *u2)
  for i in range(n):
    for j in range(3):
      for k in range(n):
        #print(new_matrix[i][k][j])
        geometric_mean[i][j] *= new_matrix[i][k][j]
      # raising to power of 1/n
      geometric_mean[i][j] = pow(geometric_mean[i][j],1/n)
  print("\n The geomteric mean is :")    
  print(geometric_mean)


  # Calculate Fuzzy weights:
  # Step 1 : Find sum ->
  #          (l1+l2+ ...ln , m1+ m2+....mn , u1+ u2+....un)
  # Step 2 : Find Reciprocal of Sum
  #         (lmu)^-1 = (1/u, 1/m,1/l)
  sum_array = numpy.zeros(3)
  inverse_array = numpy.zeros(3)
  print("\n The sum array is:")
  print(sum_array)

  for i in range(terms):
    for j in range(n):
      sum_array[i] += geometric_mean[j][i]

  # Calcualte reciprocal->
  for i in reversed(range(terms)):
    inverse_array[(terms-1)-i] = (1.0/sum_array[i])
  print("\n The inverse array is:")
  print(inverse_array)

  # Step 3: Multiply the gemometric mean and reciprocal arrays 
  #         and store that in new array : fuzzy_array
  fuzzy_array = numpy.ones( (n, terms) )


  for i in range(n):
    for j in range(terms):
      fuzzy_array[i][j] = geometric_mean[i][j]*inverse_array[j]
  print("\nThe fuzzy array is:")
  print(fuzzy_array)

  # Step 4: Defuzzification : To get Crisp numerical values. 
  # Center of Area => Wi = (l+m+u)/3

  crisp_array = numpy.zeros(n)
  total = 0
  for i in range(n):
    for j in range(terms):
      crisp_array[i] += (fuzzy_array[i][j]) 
    crisp_array[i] /= terms
    total += crisp_array[i]
  print("Total is",total)


   # Step 4: Check whether the total is greater than 1. 
   #         if it is >1 ---> Normalization
  if total >1:
    for i in range(n):
      crisp_array[i] /= total
    print("The normalized weights are", crisp_array)
  return crisp_array
```

```python
def second_step(weight):
  print("Hello", weight)
  print(len(weight))
  w = weight
  
  # columns / factors / attribute
  names = ['temperature','pressure','humidity','wind_speed','Alfalfa & Hay_percentage','Deciduous Forest_percentage','Evergreen Forest_percentage',	'Barren_percentage','Grassland_percentage',
                      'Wetlands_percentage','max_elevation','density']
  # identifying benefit-> MAX , cost -> MIN
  # we want to maximuze the attribute
  #criteria = [MAX, MAX , MAX, MAX]
  # Temp, RH , wind for area A , B , C and D

  mtx = matrix
  #mtx = [[8.2 , 51 , 6.7, 44],[18, 33, 0.9, 10],[14.6, 33, 1.3, 2],[8.3 , 97 , 4, 3],[8.3 , 97 , 4, 3]]
  #place = city names retrieved in above steps
  Data = pd.DataFrame(mtx, index = place)
  print("\n",Data)

  # vector normalization
  # for each column
  # data / sqare root (summation (sqare (data)))
  Data_norm = Data/numpy.sqrt(numpy.power(Data, 2).sum(axis = 0))
  print("\n",Data_norm)


  # calcualte weighted normalized rating
  Data_norm_w = Data_norm *w

  print("The weighted normalized \n", Data_norm_w)

  # identify postive ideal and negative ideal solution
  # for each column, we will find out the maximum and minimum
  positive_ideal = Data_norm_w.max()
  negative_ideal = Data_norm_w.min()

  best = []
  worst = []


  best = [positive_ideal[0], negative_ideal[1], negative_ideal[2], positive_ideal[3],positive_ideal[4],positive_ideal[5],
          positive_ideal[6],negative_ideal[7],positive_ideal[8],negative_ideal[9],positive_ideal[10],
          positive_ideal[11]]
  worst = [negative_ideal[0], positive_ideal[1], positive_ideal[2], negative_ideal[3],negative_ideal[4],negative_ideal[5],
           negative_ideal[6],positive_ideal[7],negative_ideal[8],positive_ideal[9],negative_ideal[10],
           negative_ideal[11]] 



  #calculate separation mesasurement
  # the distance b/w alternatives can be measured by 
  #the n-dimensional Euclidean distance. 

  # postive ideal
  Pos = numpy.sqrt(numpy.power(Data_norm_w - best , 2).sum (axis = 1))
  #print("pos\n",Pos)

  #Negative ideal
  Neg = numpy.sqrt(numpy.power(Data_norm_w - worst , 2).sum (axis = 1))
  #print("Neg\n",Neg)

  #calclate similarity to postive ideal solution
  sol = Neg/(Neg + Pos)

  # Rank the solution
  ans = sol.rank(ascending = False)

  # creating an dataframe to store the result

  temp_store = pd.DataFrame(ans)
  temp_store.columns = ['Rank']

  another = pd.DataFrame(temp_store, index = place)
  another.reset_index(inplace=True)
  another = another.rename(columns = {'index':'City'})

  another_sort = another.sort_values('Rank')
  #send_alerts()
  return another_sort



```

```python

def layer():
  l = int(input("Enter number of criteria:"))
  # Criteria Weight Calculation: 
  #l = 2
  # Creating default matrix of ones
  A = numpy.ones([l,l])
  # Running a for loop to take input from user and populate the upper triangular elements
  for i in range(0,l):
    for j in range(0,l):
      if i<j:
        aij = input('How important is option{} over option{} ?: '.format(i,j))
        A[i,j] = float(aij) #Upper triangular elements
        A[j,i] = 1/float(aij) #Lower triangular elements
  criteria = A
  x = AHP_MCDM(criteria)
  #print("This is criteria\n", x)
  Criteria_weight = fuzzy_GM(x)
  return Criteria_weight
```
